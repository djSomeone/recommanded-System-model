```python
import json
```

```python
def transform(x):
    convertedList=json.loads(x)
    return str([str(item["name"]).lower().replace(" ","") for item in convertedList]).replace("[","").replace("]","")
```

```python
df["genres"]=df["genres"].apply(transform)
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
temp=CountVectorizer()

(temp.fit_transform(df["genres"].to_list()))
```

```python
temp.get_feature_names_out()
```

```python
temp=pd.DataFrame(temp.fit_transform(df["genres"].to_list()).toarray(),columns=temp.get_feature_names_out())
```

```python
temp
```

```python
df=df.join(temp,how="outer",sort=False).drop(columns=["genres"])
```

```python
def recommadMovie(id):
    target=[df.loc[df["id"]==id].squeeze().drop(["title","id"])]
    # print((target))
    sampleSpace=df[df["id"]!=id]
    # print(sampleSpace)
    temp=[]
    for i,x in sampleSpace.iterrows():
        temp.append({"name":x["title"],"score":float(cosine_similarity(target,[x.drop(["title","id"])])[0][0])})
    
    print(sorted(temp,key=lambda x:x["score"],reverse=True)[:11])
```
